```python
duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)

df = df.drop_duplicates()

print("Shape After Removing Duplicates:", df.shape)
```

```python
df['Exercise_Per_Week'].value_counts()
```

```python
numeric_bounds = {
    'Age':                    (18, 80),
    'Work_Hours_Per_Week':    (0, 100),
    'Sleep_Hours_Night':      (3, 12),
    'Caffeine_Drinks_Day':    (0, 15),
    'Screen_Time_Hours_Day':  (0, 24),
    'Social_Media_Hours_Day': (0, 24),
    'Hobby_Time_Hours_Week':  (0, 80),
    'Close_Friends_Count':    (0, 14),
}

for col, (lo, hi) in numeric_bounds.items():
    mask = ~df[col].between(lo, hi)
    if mask.sum() > 0:
        print(f"⚠️  {col}: {mask.sum()} out of range [{lo}, {hi}]")
        print(df.loc[mask, col].value_counts())
        df.loc[mask, col] = np.nan
```

```python
allowed_values = {
    'Gender':            ['Male', 'Female', 'Other'],
    'Income_Level':      ['Low', 'Middle', 'High'],
    'Marital_Status':    ['Single', 'Married', 'Divorced', 'Widowed'],
    'Remote_Work':       ['Yes', 'No', 'Hybrid'],
    'Alcohol_Frequency': ['Never', 'Rarely', 'Sometimes', 'Often', 'Weekly', 'Daily'],
    'Diet_Quality':      ['Poor', 'Average', 'Good', 'Excellent'],
    'Smoking':           ['Never', 'Former', 'Current'],
    'Company_Mental_Health_Support': ['Yes', 'No', 'Not sure'],
    'Discuss_Mental_Health': ['Sometimes' , 'Rarely' , 'Yes easily' , 'Never'] ,
    #########################################################################################
    'Job_Satisfaction':          list(range(1, 11)),
    'Work_Stress_Level':         list(range(1, 11)),
    'Work_Life_Balance':         list(range(1, 11)),
    'Financial_Stress':          list(range(1, 11)),
    'Social_Support':            list(range(1, 11)),
    'Feeling_Sad_Down':          list(range(0, 11)),
    'Loss_Of_Interest':          list(range(0, 11)),
    'Sleep_Trouble':             list(range(0, 11)),
    'Fatigue':                   list(range(0, 11)),
    'Poor_Appetite_Or_Overeating': list(range(0, 11)),
    'Feeling_Worthless':         list(range(0, 11)),
    'Concentration_Difficulty':  list(range(0, 11)),
    'Anxious_Nervous':           list(range(0, 11)),
    'Mood_Swings':               list(range(0, 11)),
    'Irritability':              list(range(0, 11)),
    'Obsessive_Thoughts':        list(range(0, 11)),
    'Compulsive_Behavior':       list(range(0, 11)),
    'Panic_Attacks':             list(range(0, 5)),
    'Feel_Understood':           list(range(0, 11)),
    'Loneliness':                list(range(0, 11)),


}


for col, allowed in allowed_values.items():
    if col not in df.columns:
        continue
    invalid = ~df[col].isin(allowed + [np.nan])
    if invalid.sum() > 0:
        print(f"⚠️  {col} Unexpected Values: ")
        print(df.loc[invalid, col].value_counts())
        df.loc[invalid, col] = np.nan
```

```python
df['Gender'].value_counts(dropna=False)
```

```python
binary_cols = [
    'Ever_Bullied_At_Work',
    'Self_Harm_Thoughts',
    'Suicidal_Thoughts',
    'Family_History_Mental_Illness',
    'Previously_Diagnosed',
    'Ever_Sought_Treatment',
    'On_Therapy_Now',
    'On_Medication',
    'Trauma_History',
    'Has_Mental_Health_Issue',
]

for col in binary_cols:
    if col not in df.columns:
        continue
    invalid = ~df[col].isin(['yes', 'no', 0 , 1 , np.nan])
    if invalid.sum() > 0:
        print(f"⚠️  {col} Unexpected Values:")
        print(df.loc[invalid, col].value_counts())
        df.loc[invalid, col] = np.nan
```

```python
null_summary = df.isnull().sum()
null_percent = (df.isnull().sum() / len(df) * 100).round(2)

null_report = pd.DataFrame({
    'Missing Count': null_summary,
    'Missing %': null_percent
}).query('`Missing Count` > 0').sort_values('Missing %', ascending=False)

print(null_report)
```

```python
#Logical Consistency

inconsistent = df[(df['On_Therapy_Now'] == 1) & (df['Ever_Sought_Treatment'] == 0)]
print(f"Therapy without Treatment: {len(inconsistent)}")

inconsistent2 = df[(df['On_Medication'] == 1) & (df['Previously_Diagnosed'] == 0)]
print(f"Medication Without Diagnosis: {len(inconsistent2)}")
```

```python
from sklearn.impute import SimpleImputer

gender_imputer = SimpleImputer(strategy='most_frequent')

df['Gender'] = gender_imputer.fit_transform(
    df[['Gender']]
).ravel()
```
