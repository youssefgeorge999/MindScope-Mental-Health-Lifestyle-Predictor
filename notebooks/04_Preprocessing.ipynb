```python
import pandas as pd
from sklearn.preprocessing import (
    MinMaxScaler,
    StandardScaler,
    RobustScaler
)
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('/content/final_featured_dataset.csv')
```

```python
df['Gender_Male'] = df['Gender_Male'].astype(int)
```

```python
X = df.drop(
    'Has_Mental_Health_Issue',
    axis=1
)

y = df['Has_Mental_Health_Issue']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
```

```python
print(X_train.shape)
print(X_test.shape)

print(y_train.shape)
print(y_test.shape)
```

```python
print(y_train.value_counts(normalize=True))

print()

print(y_test.value_counts())
```

```python
minmax_cols = [
    'Mood_Swings',
    'Concentration_Difficulty',
    'Fatigue',
    'Anxious_Nervous',
    'Financial_Stress',
    'Feeling_Worthless',
    'Panic_Attacks',
    'Poor_Appetite_Or_Overeating',
    'Irritability',
    'Work_Stress_Level',
    'Discuss_Mental_Health',
    'Sleep_Trouble',
    'Income_Level',
    'Loneliness',
    'Compulsive_Behavior',
    'Work_Life_Balance',
    'Feeling_Sad_Down',
    'Diet_Quality',
    'Loss_Of_Interest',
    'Alcohol_Frequency',
    'Social_Support',
    'Feel_Understood',
    'Job_Satisfaction',
    'Obsessive_Thoughts'
]

score_cols = [
    'Stress_Burden_Score',
    'Exhaustion_Score',
    'Emotional_Distress_Score',
    'Cognitive_Difficulty_Score',
    'Social_Isolation_Score'
]

continuous_cols = [
    'Age',
    'Caffeine_Drinks_Day',
    'Close_Friends_Count',
    'Hobby_Time_Hours_Week',
    'Screen_Time_Hours_Day',
    'Social_Media_Hours_Day',
    'Work_Hours_Per_Week',
    'Sleep_Hours_Night',
    'Lifestyle_Imbalance',
    'Stress_x_Loneliness',
    'Stress_x_Sleep',
    'SocialMedia_x_Sleep'
]
```

```python
minmax_scaler = MinMaxScaler()

robust_scaler = RobustScaler()
```

```python
X_train[minmax_cols] = minmax_scaler.fit_transform(
    X_train[minmax_cols]
)

X_train[continuous_cols] = robust_scaler.fit_transform(
    X_train[continuous_cols]
)
```

```python
X_test[minmax_cols] = minmax_scaler.transform(
    X_test[minmax_cols]
)

X_test[continuous_cols] = robust_scaler.transform(
    X_test[continuous_cols]
)
```

```python
X_train
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(
    random_state=42
)
```

```python
X_train_resampled, y_train_resampled = smote.fit_resample(
    X_train,
    y_train
)
```

```python
print(y_train.value_counts())

print()

print(y_train_resampled.value_counts())
```

```python
print(X_train.shape)
print(X_train_resampled.shape)

print()

print(y_train.shape)
print(y_train_resampled.shape)
```

```python
train_processed_df = X_train_resampled.copy()

train_processed_df['Has_Mental_Health_Issue'] = y_train_resampled
```

```python
train_processed_df.to_csv(
    '/content/Train_Preprocessed_Data.csv',
    index=False
)
```

```python
test_processed_df = X_test.copy()

test_processed_df['Has_Mental_Health_Issue'] = y_test
```

```python
test_processed_df.to_csv(
    '/content/Test_Preprocessed_Data.csv',
    index=False
)
```
